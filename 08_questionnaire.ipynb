**Question 1: What problem does collaborative filtering solve?**

Collaborative filtering is a recommender system technique. In general, it recommends items that were bought by users that liked similar items to what the user bought.

**Question 2: How does it solve it?**

* Look at which products the current user has used or liked
* Find other users who have liked or used similar proudcts
* Recommend other products that those users have used or liked

**Question 3: Why might a collaborative filtering predictive model fail to be a very useful recommendation system?**

The main issue with this approach is that it may recommend items that are not interesting to the user but that are consider by the majority of the user base. It also has an scalability problem (in its default form) and a cold-start problem.

**Question 6: What is a latent factor? Why is it "latent"?**

The latent factor is an unobserved vector that represents the "type" of user and the "type" of item. Each dimension of the latent factor represents one characteristic. It is latent because those characteristics are learnt from the data but do not specifically correspond to observed attributes.

**Question 7: What is a dot product? Calculate a dot product manually using pure Python with lists.**

The dot product is a special case of matrix multiplication when the first matrix is a nx1 and the second one is a 1xn matrix. It returns a single number.

```python
a = [2,3,4]
b = [3,3,5]

# Python List
res = 0
for i in range(0,len(a)):
    res += a[i]*b[i]
print(res)

#PyTorch (2 ways)
a_t = torch.tensor(a)
b_t = torch.tensor(b)
print(a_t @ b_t)
print(torch.dot(a_t, b_t))
```

**Question 9: What is an embedding matrix?**

It considers all the latent factors of users or items. Each user/item has a vector (i.e., latent factor). The combination of all the latent factors generates a matrix.

**Question 10: What is the relationship between an embedding and a matrix of one-hot-encoded vectors?**

To calculate the result for a particular item and user combination, we have to look up the index of the movie in our movie latent factor matrix, and the index of the user in our user latent factor matrix. However, looking up an index is not an operation that deep learning frameworks can parallelize. 

Fortunately, we can represent "looking up an index" as a matrix product. The idea is to replace the indices with a one-hot encoded matrix. This way, we can estimate the predicted value by simply doing a matrix multiplication operation between the corresponding user and item latent factors.

<span style="color:blue">**Embedding is usually referred to the latent factor matrix**.</span>

```python
class DotProduct(Module):
    def __init__(self, n_users, n_movies, n_factors):
        self.user_factors = Embedding(n_users, n_factors)
        self.movie_factors = Embedding(n_movies, n_factors)
        
    def forward(self, x):
        users = self.user_factors(x[:,0])
        movies = self.movie_factors(x[:,1])
        return (users * movies).sum(dim=1)

ratings = pd.read_csv(path/'u.data', delimiter='\t', header=None,
                      names=['user','movie','rating','timestamp'])

dls = CollabDataLoaders.from_df(ratings[["user", "movie", "rating"]], bs=64)
dls.show_batch()

n_users  = len(dls.classes['user'])
n_movies = len(dls.classes['movie'])
model = DotProduct(n_users, n_movies, 50)
learn = Learner(dls, model, loss_func=MSELossFlat())
learn.fit_one_cycle(5, 5e-3)
```
